```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv("Lionel Messi Goals.csv")
data.head()
```

```python
df = data["GoalType"].value_counts().reset_index()
df
```

```python
plt.pie(df["GoalType"], labels=df["index"])
plt.show()
```

```python
df = data["Goal-Method"].value_counts().reset_index()
df
```

```python
plt.bar(df["index"], df["Goal-Method"])
plt.show()
```

```python
data["Minute"] = data["Minute"].apply(eval)
```

```python
data["Minute"] = data["Minute"].astype(int)
```

```python
df = data["Minute"].value_counts().reset_index().sort_values(by=["index"], ascending=True)
df
```

```python
plt.figure(figsize=(24, 8))
plt.plot(df["index"], df["Minute"])
plt.show()
```
