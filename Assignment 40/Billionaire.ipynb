```python
import pandas as pd
import matplotlib.pyplot as plt
import statistics
```

```python
data = pd.read_csv("Billionaire.csv")
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
data["NetWorth"] = data["NetWorth"].str.strip("$")
data["NetWorth"] = data["NetWorth"].str.strip("B")
data["NetWorth"] = data["NetWorth"].astype(float)
```

```python
data.head()
```

```python
df = data.sort_values(by = ["NetWorth"], ascending=False).head(10).reset_index()
df
```

```python
plt.figure(figsize= (24,8))
plt.bar(df["Name"], df["NetWorth"])
plt.show()
```

```python
df = data["Source"].value_counts().head(10).reset_index()
df
```

```python
plt.pie(df["count"], labels=df["Source"])
plt.show()
```

```python
df = data["Industry"].value_counts().reset_index()
df
```

```python
plt.figure(figsize=(24,14))
plt.pie(df["count"], labels = df["Industry"])
plt.show()
```

```python
df = data["Country"].value_counts().head().reset_index()
df
```

```python
plt.pie(df["count"], labels = df["Country"])
plt.show()
```

```python
df = data["Age"].mean()
df
```

```python
ages = []
for i in data["Age"]:
    ages.append(i)

ages
```

```python
statistics.variance(ages)
```

```python
df = data.groupby("Country")["NetWorth"].sum().reset_index()
df = df.sort_values(by='NetWorth')

plt.figure(figsize=(12,6))
plt.plot(df["Country"],df["NetWorth"])
plt.xticks(rotation=90)
plt.show()
```

```python
df=data.query('Country == "United States"').reset_index()
df.head(10)
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.head(10)['Name'],df.head(10)['NetWorth'])
plt.xticks(rotation=90)
plt.show()
```

```python
df = data["Country"].value_counts().reset_index()
df = df.sort_values(by="count",ascending=True)
df.head(10)
```

```python
plt.figure(figsize=(12,6))
plt.plot(df.head(10)['Country'],df.head(10)['count'])
plt.xticks(rotation=90)
plt.show()
```
