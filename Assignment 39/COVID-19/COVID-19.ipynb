```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tx_data = pd.read_csv("input/COVID-19-time-series-clean-complete.csv")

tx_data.head()
```

```python
tx_data_1 = tx_data.query("Date >= '2020-04-01' and Date < '2020-05-01'").reset_index(drop=True)

tx_data_1.head(10)
```

```python
tx_data_2 = tx_data_1.groupby('Country')['Active'].mean().reset_index()
tx_data_2 = tx_data_2.sort_values(by="Active", ascending=False).reset_index()

tx_data_2.head(8)
```

```python
data_iran = tx_data.query("Country == 'Iran'").reset_index(drop=True)
data_iran['Date'] = pd.to_datetime(data_iran['Date'])

data_iran.head(10)
```

```python
plt.figure(figsize=(12,6))

plt.plot(data_iran['Date'].to_numpy(dtype=str), np.log10(data_iran['New deaths'].to_numpy()))

plt.grid()
plt.show()
```
